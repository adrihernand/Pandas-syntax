# 6.3 Column contains multiple variables

```python
# Let's import the ebola data

ebola = pd.read_csv("C:/Users/adri_/Documents/GitHub- Adriana/Pandas for everyone/data/country_timeseries.csv")
```

```python
# let;s find out the column names


ebola.columns
```

```python
# let;s focus in these columns
ebola.iloc[:5, [0,1,2,3,10,11]]
```

```python
# Melting the columns

ebola_long = pd.melt(ebola, id_vars = ["Date", "Day"])
ebola_long
```

# 6.3.1 Split and add columns individually(Simple Method)

```python
# Get the variable column
# access the string method
# and split the column based on a delimiter

variable_split = ebola_long.variable.str.split("_")
variable_split
```

```python
# Let's see te entire container

type(variable_split)
```

```python
# Use the "get" method to get index of each row

print(pd.melt(ebola, id_vars = ["Date", "Day"]).variable.str.split("_").str.get(0))
print(pd.melt(ebola, id_vars = ["Date", "Day"]).variable.str.split("_").str.get(1))
```

# 6.3.2 Split and Combine in a single step (Simple method)

```python
#Split and combine

variable_split = ebola_long.variable.str.split("_", expand = True)
variable_split.columns = ["status", "country"]
ebola_parsed = pd.concat([ebola_long, variable_split], axis = 1)

ebola_parsed
```

# 6.3.3 Split and combine in a single step (Complicated method)

```python
# We can combine a list of split items with the zip function

constants = ["pi", "e"]
values = ["3.14", "2.718"]

print(list(zip(constants,values)))
```

```python
# Unpacking a container

ebola_long["status"],ebola_long["country"] = zip(*ebola_long.variable.str.split("_"))
ebola_long
```

# 6.4 Variables in both rows and columns

```python
# Let's import the data

weather = pd.read_csv("C:/Users/adri_/Documents/GitHub- Adriana/Pandas for everyone/data/weather.csv")
weather
```

```python
# Let's unpivot(melt)

weather_melt = pd.melt(weather, id_vars = ["id", "year", "month", "element"],
                      var_name = "day",
                      value_name = "temp")
weather_melt
```

```python
# Doing casting (or spreading)
weather_tidy = weather_melt.pivot_table(
    index= ["id", "year", "month", "day"],
    columns = "element",
    values = "temp")
```

```python
# Flattening the table
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat 
```

# 6.5 Multiple observational data

```python
# Let's look at the billboard data
billboard
```

```python
#Melting columns

billboard_long = pd.melt(billboard, id_vars = ["year","artist","track","time","date.entered"], var_name = "week", value_name= "rating")
billboard_long

```

```python
#Subset based on track

billboard_long[billboard_long.track == "Loser"]
```

```python
#Placing year, artisi, track and time in a new dataframe

billboard_songs = billboard_long[["year", "artist", "track", "time"]]
billboard_songs.shape
```

```python
# Drop duplicate entries
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs.shape
```

```python
# Assigning each row a unique value
billboard_songs["id"] = range(len(billboard_songs))
billboard_songs
```

```python
# Merge the song dataframe and the original data set
billboard_ratings = billboard_long.merge(
    billboard_songs, on = ["year", "artist", "track", "time"])

billboard_ratings.shape

```

```python
# Subset the columns to the ones we want in the ratings dataframe
billboard_ratings = billboard_ratings[["id", "date.entered", "week", "rating"]]
billboard_ratings
```
