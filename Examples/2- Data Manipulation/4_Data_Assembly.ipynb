# 4.4.1 One to One merge

```python
# I would like to merge the "site" file with the "visited" file.

#I will modify the visited table to a table with only 3 values, so each one match the "Site" table.

visited_subset = visited.loc[[0,2,6],]
visited_subset
```

```python
# One to one merging

o2o_merge = site.merge(visited_subset, left_on= "name", right_on="site")

o2o_merge
```

# 4.4.2 Many to one merge

```python
# I would like to merge the "site" file with the "visited" file but I won't modify anything, so if "visited" has repetitions
# in the "site" columns, that would be kept and the resulting dataframe will have repetitions.


# Merging many to one
m2o_merge = site.merge(visited, left_on = "name", right_on ="site" )
m2o_merge
```

# 4.4.3 Many to many

```python
#I would like to merge dataframes that came from mergin other dataframes. In this case, the table "person" was previously 
# merged with the table "survey"  and the table "visited" was previously merged with the table "survey" 


ps = person.merge(survey, left_on = "ident", right_on = "person")
vs = visited.merge(survey, left_on = "ident", right_on = "taken" )
```

```python
ps
```

```python
vs
```

```python
#Merging  many to many



ps_vs = ps.merge(vs,left_on=["ident", "taken", "quant", "reading"], right_on=["person","ident","quant","reading"])
ps_vs
```
