# Chapter 3 - Introduction to Plotting

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the data for Anscombe's quartet
anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
# Subsetting the data to plot (I want to plot only the datatset 1) 

dataset_1 = anscombe[anscombe["dataset"]=="I"]

plt.plot(dataset_1["x"],dataset_1["y"])
```

```python
# Subsetting the data to plot (I want to plot only the datatset 1 using points) 

dataset_1 = anscombe[anscombe["dataset"]=="I"]

plt.plot(dataset_1["x"],dataset_1["y"],"o")
```

```python
#Renaming subsets for easy manipulation


dataset_2 = anscombe[anscombe["dataset"]=="II"]

dataset_3 = anscombe[anscombe["dataset"]=="III"]

dataset_4 = anscombe[anscombe["dataset"]=="IV"]
```

```python
# Creating a plot of 4 sub-plots (2 by 2)


fig = plt.figure()       #Creates the big plot

axes1 = fig.add_subplot(2, 2, 1)  #plot location 1
axes2 = fig.add_subplot(2, 2, 2)  #plot location 2
axes3 = fig.add_subplot(2, 2, 3)  #plot location 3
axes4 = fig.add_subplot(2, 2, 4)  #plot location 4


# Plotting all using the "axes" which is diff than axis. Plotting the data
axes1.plot(dataset_1["x"],dataset_1["y"],"o")    
axes2.plot(dataset_2["x"],dataset_2["y"],"o")
axes3.plot(dataset_3["x"],dataset_3["y"],"o")
axes4.plot(dataset_4["x"],dataset_4["y"],"o")


axes1.set_title("dataset 1")                    # Assign names on plots
axes2.set_title("dataset 2")
axes3.set_title("dataset 3")
axes4.set_title("dataset 4")


fig.suptitle("Anscombe data")                   #Assign a name to the plot


fig.tight_layout()                              #Use a fugure layout



axes1.set_title("dataset 1")                    # Assign names on plots
axes2.set_title("dataset 2")
axes3.set_title("dataset 3")
axes4.set_title("dataset 4")



```

# 3.3 Statistical graphics using Matplotlib

```python
# Loading the data

tips = sns.load_dataset("tips")
tips
```

# 3.3.1 Univariate

# 3.3.1.1 Histograms

```python
# Histograms

fig = plt.figure()                                                  # Defining the plot name


# Data to plot
axes1 = fig.add_subplot(1,1,1)
axes1.hist(tips["total_bill"], bins = 10)                             


axes1.set_title("Histogram of Total Bill")
axes1.set_xlabel("Frequency")
axes1.set_ylabel("Total Bill")

```

# 3.3.2 Bivariate

# 3.3.2.1 Scatter plot

```python
# Scatter plots

scatter_plot = plt.figure()                                             # Defining the plot name



 # Data to plot
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips["total_bill"],tips["tip"])                            


axes1.set_title("Scatter plot of Total Bill vs Tips")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tips")


```

# 3.3.2.2 Boxplots

```python
# Boxplots

boxplot = plt.figure()                                             # Defining the plot name


# Data to plot 
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot(
    [tips[tips["sex"] == "Female"]["tip"],
     tips[tips["sex"] == "Male"]["tip"]],
    labels = ["Female","Male"])                                                           


axes1.set_title("Boxplot of tips  by sex")
axes1.set_xlabel("Sex")
axes1.set_ylabel("Tip")


```

# 3.3.3 Multivariate Data

```python
# Scatter plots multivariate

def recode_sex(sex):
    if sex == "Female":
        return 0
    else:
        return 1
    
tips["sex_color"] = tips["sex"].apply(recode_sex)


scatter_plot = plt.figure()                                             # Defining the plot name



# Data to plot
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x= tips["total_bill"],
    y= tips["tip"],
    s = tips["size"],
    c = tips["sex_color"],
    alpha = 0.5)                                                       


axes1.set_title("Scatter plot of Total Bill vs Tips by Sex and Size by Size")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tips")

```

# 3.4.2 Bivariate data

# 3.4.2.1 Scatter plot

```python
# Defining the plot name
scatter, ax = plt.subplots()


# Data to plot using regplot
ax = sns.regplot(x = "total_bill", y = "tip",  data = tips)
ax.set_title("Scatterplot of Total Bills and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")


########################################################################

# Defining the plot name  ----------------------------- Only the scatter
scatter, ax = plt.subplots()


# Data to plot using regplot
ax = sns.regplot(x = "total_bill", y = "tip",  data = tips, fit_reg = False)
ax.set_title("Scatterplot of Total Bills and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")


########################################################################

# Data to plot using lmplot
fig = sns.lmplot(x = "total_bill", y = "tip",  data = tips)


########################################################################

# Data to plot using a Joint Grid object
joint = sns.jointplot(x = "total_bill", y = "tip",  data = tips)
joint.set_axis_labels(xlabel = "Total Bill", ylabel ="Tip")
joint.fig.suptitle("Joint Plot of Total Bill and Tip", fontsize = 10, y=1.03)

```

# 3.4.2.3 2D Density Plot

```python
# Defining the plot name  ----------------------------- density plot across 2 variables
kde, ax = plt.subplots()


# Data to plot using regplot
ax = sns.kdeplot(data = tips["total_bill"],data2 = tips["tip"], shade = True )
ax.set_title("Kernel density plot of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")

##############################################################################################

# ----------------------------- density plot across 2 variables using jointplot


# Data to plot using lmplot
kde_jointplot = sns.jointplot(x = "total_bill", y = "tip",  data = tips,kind ="kde")

ax.set_title("Kernel density plot of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
```

# 3.6 Seaborn Themes and Styles

```python
 # Initial plot for comparison

fig, ax = plt.subplots()
ax = sns.violinplot(x = "time", y = "total_bill", hue = "sex", data = tips, split = True)
```

```python

# set style and plot                                         # Style example
sns.set_style('darkgrid')


fig, ax = plt.subplots()
ax = sns.violinplot(x = "time", y = "total_bill", hue = "sex", data = tips, split = True)
```

```python

# set style and plot                                         # Style example
sns.set_style('whitegrid')


fig, ax = plt.subplots()
ax = sns.violinplot(x = "time", y = "total_bill", hue = "sex", data = tips, split = True)
```

```python

# set style and plot                                         # All Styles example 
fig = plt.figure()
seaborn_styles = ["darkgrid", "whitegrid","dark", "white","ticks"]
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2,3,plot_position)
        violin = sns.violinplot(x = "time", y = "total_bill", data= tips, ax = ax)
        violin.set_title(style)
        fig.tight_layout()
    
```
